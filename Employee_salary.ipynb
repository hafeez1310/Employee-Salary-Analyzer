```python
import numpy as np
```

```python
names = np.array([
    'Ali', 'Sara', 'John', 'Aisha',"David",
    "Fatima","Rahul","Zoya","Kiran","Ahmed"
])

department = np.array([
    'IT', 'HR', 'IT', 'Finance', 'Marketing',"HR","IT","Finance","Marketing","IT"
])

salary = np.array([
    55000,45000,75000,82000,50000,48000,39000,91000,52000,88000
])

experience = np.array([
    2,1,5,7,3,2,1,8,4,6
])

rating = np.array([
    4,2,5,5,4,4,2,5,3,5
])
```

Total Names

```python
print(len(names))
```

Average Salary

```python
avg_salary = np.mean(salary)
print(avg_salary)
```

Highest Salarym

```python
highest_salary = np.argmax(salary)
print(names[highest_salary])
print(salary[highest_salary])
```

Lowest Salarym

```python
lowest_salary = np.argmin(salary)
print(names[lowest_salary])
print(salary[lowest_salary])
```

IT Employee

```python
it_emp = department == 'IT'
it_avg_salary = np.mean(salary[it_emp])
print(it_avg_salary)
```

HR Employee

```python
hr_emp = department == 'HR'
print(names[hr_emp])
```

Employee with more than five years of Experience

```python
senior_emp = experience > 5
print(names[senior_emp])
print(experience[senior_emp])
```

Average Salary of Experienced Employees


```python
experience_avg_salary = np.mean(experience[senior_emp])
print(experience_avg_salary)
```

Bonus Calculation

Company gives:

20% bonus if rating = 5
10% bonus otherwise

```python
bonus = np.where(rating == 5, salary * 0.20, salary * 0.10 )
print(bonus)
```

Final Salary After Bonus

```python
final_salary = salary + bonus
print(final_salary)
```

Promotion Eligibility

```python
promotion = (experience>=5) & (rating>=4)
print(names[promotion])
```

Sort Employees by Salary Descending

```python
sorted_index = np.argsort(salary)[::-1]
print(names[sorted_index])
```

Top 3 Highest Paid Employees

```python
print(names[sorted_index][:3])
```

Employees Earning More Than Average

```python
avg_earning = np.mean(salary)
highest_paid = salary > avg_earning
print(names[highest_paid])
```

Save Report to CSV

```python
report = np.column_stack((
    names,
    department,
    salary,
    experience,
    rating,
    final_salary
))

np.savetxt(
    "employee_report.csv",
    report,
    delimiter=',',
    fmt='%s',
    header='Names, Department, Salary, Experience, Rating, Final Salary',
    comments=''
)

print('Report Saved')
```

Find Department with highest average salary

```python
it_dept_avg =np.mean( salary[department == 'IT'])
hr_dept_avg =np.mean( salary[department == 'HR'])
finance_dept_avg =np.mean( salary[department == 'Finance'])
marketing_dept_avg =np.mean(salary[department == 'Marketing'])

print(it_dept_avg)
print(hr_dept_avg)
print(finance_dept_avg)
print(marketing_dept_avg)

avg_list = np.array([
    it_dept_avg,
    hr_dept_avg,
    finance_dept_avg,
    marketing_dept_avg
])
dept_names = np.array(['IT', 'HR', 'Finance', 'Marketing'])


dept_highest_avg = dept_names[np.argmax(avg_list)]
print('Department with Highest Average is :', dept_highest_avg)
```
